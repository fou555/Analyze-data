```python
def print_basic_statistics(dataframe):
    print(dataframe.sum(axis=0))
    print(dataframe.mean(axis=0))
    print(dataframe.median(axis=0))
    print(dataframe.min(axis=0))
    print(dataframe.max(axis=0))
    print(dataframe.std(axis=0))
print_basic_statistics(airtravel)
```

```python
employee = pd.read_csv('employee_data.csv')
print(employee)
```

```python
employee.rename(columns={'number': 'staff_id'}, inplace=True) #เปลี่ยนชื่อ column
print(employee.describe(include='all')) #ดูค่าสถิติเบื้องต้น
#print(employee.head())
```

```python
print(employee.isnull().sum()) #ดูว่าแต่ละ column มีค่าว่างเท่าไหร่
```

```python
employee = employee.drop("tax_file_no",axis=1) #ลบ column tax_file_no ทิ้งเนื่องจากไม่มีข้อมูล
print(employee)
```

```python
#จัดการกับแถวที่มีค่าว่างเยอะ โดยเราจะลบเมื่อแถวนั้นมีข้อมูลไม่ถึง 2 คอลัมน์
#column ที่จะลบ
under_threshold_removed = employee.dropna(axis='index', thresh=2, inplace=False)
under_threshold_rows = employee[~employee.index.isin(under_threshold_removed.index)]
print(under_threshold_rows)
```

```python
employee.dropna(axis='index', thresh=2, inplace=True)
print(employee)
```

```python
# เนื่องจากเพศมีค่าว่าง เราจึงแทนที่ค่าว่างนั้นด้วย U
employee['gender'] = employee['gender'].fillna('U')
print(employee)
```

```python
print(employee.isnull().sum())
```

```python
# เอาข้อมูลที่ทำความสะอาดแล้วส่งออกเป็น .csv
employee.to_csv(index=False)
```

```python
svday_temp = { "day" : ["Sun","Mon","Tue","Wed","Thu","Fri","Sat"],
          "temp" : [6,8,6,8,10,8,6]
        }
df = pd.DataFrame(svday_temp) #สร้าง dataframe จากข้อมูล

#ให้แกน x คือวัน แกน y คืออุณหภูมิ
x = df["day"]
y = df["temp"]
plt.title("Temperature in 7 days")
plt.plot(x,y, ls="--", marker="o", color="m") #สั่งให้วาดกราฟ
plt.show()  #แสดงผลกราฟออกมา
```

```python
sales = {
    "month" : ["Jan", "Feb" , "Mar"],
    "sales" : [20000, 25000 , 90000]
}
df_sales = pd.DataFrame(sales) #สร้าง dataframe จากข้อมูล
x = df_sales["month"] #ให้แกน x เป็น เดือน
y = df_sales["sales"] #ให้แกน y เป็น ยอดขาย
plt.title("3 Month Sale")
plt.bar(x,y, color=["c","m","y"], ls="-.", hatch="xx")
plt.show()
```

```python
score = { "math_score" : [40,45,50,65,60],
          "science_score" : [60,70,80,90,100],
          "eng_score" : [50,60,75,78,88],
          "thai_score" : [79,70,75,80,72] }
df_score = pd.DataFrame(score)
graph_color = ["yellow","green","blue","pink"]
bplot = plt.boxplot(df_score,widths=0.2, labels=["Math","Sci","Eng","Thai"], patch_artist=True)
for patch, color in zip(bplot["boxes"], graph_color):
  patch.set_facecolor(color)
plt.show()
```
